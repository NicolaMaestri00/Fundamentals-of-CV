```python
import numpy as np

img_R = cv2.imread('data/00351v_R.jpg')
img_G = cv2.imread('data/00351v_G.jpg')
img_B = cv2.imread('data/00351v_B.jpg')

gray_R = cv2.cvtColor(img_R, cv2.COLOR_BGR2GRAY)
gray_G = cv2.cvtColor(img_G, cv2.COLOR_BGR2GRAY)
gray_B = cv2.cvtColor(img_B, cv2.COLOR_BGR2GRAY)
img_not_aligned = np.stack([gray_R, gray_G, gray_B], axis=2)
img_aligned = channels.align(gray_R, gray_G, gray_B)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img_not_aligned)
plt.title('Not aligned')
plt.subplot(1, 2, 2)
plt.imshow(img_aligned)
plt.title('Aligned')
```
